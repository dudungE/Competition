# DSR, DTI 파생변수 추가
- DSR = 원리금상환액/가처분소득
- DTA = 금융부채/금융자산

- HDRI = [(1+(DSR-0.4))+(1+(DTA-1))]*100

```python
data = pd.read_csv('./data/2020가계부채/가계부채_2020_rev.csv', engine='python', encoding='cp949')
data
```

```python
data["금융자산"]
```

```python
data["금융부채"]
```

```python
data['DTA'] = data["금융부채"]/data["금융자산"]
```

```python
print(data['원리금상환액'], '\n')
print(data['처분가능소득(보완)[경상소득(보완)-비소비지출(보완)]'])
```

```python
data[data['처분가능소득(보완)[경상소득(보완)-비소비지출(보완)]']<0]
```

```python
data['DSR'] = data['원리금상환액']/data['처분가능소득(보완)[경상소득(보완)-비소비지출(보완)]']
```

```python
data['HDRI'] = ((1+data['DSR']-0.4)+(1+data['DTA']-1))*100
```

```python
data
```

```python
data.isnull().sum()
```

```python
columns=['DSR','DTA','HDRI','30일 이상 원리금 연체 여부','원리금 연체 여부','부채 증가 원인']
data1 = data[columns]
data1
```

```python
data2 =data1[(data1['원리금 연체 여부']=='있음') & (data1['HDRI']>100) & (data1['30일 이상 원리금 연체 여부']=='있음')]
data2
```

```python
x=data1[(data1['HDRI']>100) & (data1['원리금 연체 여부']=='있음')]
x['부실가구']='1'
x
# HDRI 100 이상이면서 원리금 연체 된 가구를 부실가구로 정의
```

```python
data3 = pd.merge(data1, x, how='outer')
data3
```

```python
data3 = data3.fillna({'부실가구':0})
data3
```
