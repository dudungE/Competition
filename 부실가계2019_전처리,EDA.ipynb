# 전처리

### DSR, DTI, HDRI 파생변수 추가
- DSR = 원리금상환액/가처분소득
- DTA = 금융부채/금융자산

- HDRI = [(1+(DSR-0.4))+(1+(DTA-1))]*100

```python
data = pd.read_csv('./data/2019가계부채/가계금융복지조사_2019.csv', engine='python', encoding='cp949')
# data
```

```python
# data["금융자산"]
# data["금융부채"]
```

```python
# DTA(총부채상환비율) 구하기
data['DTA'] = data["금융부채"]/data["금융자산"]
```

```python
print(data['원리금상환액'], '\n')
print(data['처분가능소득(보완)[경상소득(보완)-비소비지출(보완)]'])
```

```python
# DSR(총부채원리금상환비율) 구하기
data['DSR'] = data['원리금상환액']/data['처분가능소득(보완)[경상소득(보완)-비소비지출(보완)]']
```

```python
# HDRI 계산
data['HDRI'] = ((1+data['DSR']-0.4)+(1+data['DTA']-1))*100
```

```python
data
```

```python
# 연체여부 결측치 행 제거
x = data['원리금 연체 여부'].dropna()
len(x)

data = data.dropna(axis=0, subset=['원리금 연체 여부'])   
```

```python
data

# 데이터 개수 9733개
```

- 원리금 연체 여부있으면서 HDRI>100인 가구 부실가구로 지정

```python
# HDRI 100이상이고 원리금 연체 여부 있을 때 부실가계 파생변수 부여

data2=data[(data['HDRI']>100) & (data['원리금 연체 여부']=='있음')]
data2['부실가계']='1'
data2
# HDRI 100 이상이면서 원리금 연체 된 가구를 부실가구로 정의
# 총 944가구 결과로 나옴
```

```python
# 부실가계와 아닌가구 merge
data3 = pd.merge(data, data2, how='outer')

# NaN 0으로 채우기
data3['부실가계']=data3['부실가계'].fillna(0)

data3
```

```python
data3.isnull().sum()
```

```python
data3.to_csv("./data/2019가계부채/label_2019.csv", mode="w",encoding='euc-kr')

# 파생변수 지정 데이터 1차 저장
```

### 범주형 데이터 encoding

```python
import pandas as pd
import numpy as np
```

```python
# 데이터 reload
data = pd.read_csv('./data/2019가계부채/label_2019.csv', engine='python', encoding='cp949')
data
```

```python
# feature slicing

# 담보대출 용도 및 출처
df_1 = data.iloc[:,51:64]

# 신용대출 용도 및 출처
df_2 = data.iloc[:,75:88]
df_2

# 가구주 직업은 상용근로자 or not, 입주형태는 자가 or not
df_3 = data[['가구주 성별','가구주교육정도 통합별','가구주 혼인상태','입주형태','수도권여부','가구주 직업(대분류)','부실가계']]

df = pd.concat([df_1,df_2,df_3],axis=1)
df.info()
```

```python
# 가구주직업 인코딩(단순노무 종사자=1, or 0)
for i in range(len(df['가구주 직업(대분류)'])):
    if df['가구주 직업(대분류)'][i]=='단순노무 종사자':
        df['가구주 직업(대분류)'][i]=1
    else:
        df['가구주 직업(대분류)'][i]=0
        
# 입주형태 인코딩(자기집 =1 , or 0)
for i in range(len(df['입주형태'])):
    if df['입주형태'][i]=='자기집':
        df['입주형태'][i]=1
    else:
        df['입주형태'][i]=0
        
# 가구주 혼인상태 인코딩(배우자있음 =1 , or 0)
for i in range(len(df['가구주 혼인상태'])):
    if df['가구주 혼인상태'][i]=='배우자있음':
        df['가구주 혼인상태'][i]=1
    else:
        df['가구주 혼인상태'][i]=0
        
# 가구주교육정도 통합별 인코딩(대학졸업 이상 =1 , or 0)
for i in range(len(df['가구주교육정도 통합별'])):
    if df['가구주교육정도 통합별'][i]=='대학졸업 이상':
        df['가구주교육정도 통합별'][i]=1
    else:
        df['가구주교육정도 통합별'][i]=0
```

```python
# 나머지 binary 범주형 데이터 인코딩

from sklearn.preprocessing import LabelEncoder

for i in df.columns:
    for j in range(len(df)):
        if type(df[i][j]) == str:
            items = df[i].unique()
            encoder = LabelEncoder()
            encoder.fit(items)
            df[i] = encoder.transform(data[i])
```
